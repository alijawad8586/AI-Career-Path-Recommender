```python
from typing_extensions import TypedDict
```

```python
from langgraph.graph import StateGraph, START, END
```

```python
class StudentState(TypedDict):
    question: str
    answer: str
```

```python
def answer_node(state: StudentState):
  question = state["question"]
  question = "Explain LCEL"
  return {"answer": answer}
```

```python
## agent 2
def quiz_node(state: StudentState):
    question = state["question"]

    answer = (
        "Quiz for: " + question + "\n"
        "1. What is the main idea?\n"
        "2. Why is it useful?\n"
        "3. Give one example."
    )

    return {"answer": answer}
```

```python
def summary_node(state: StudentState):
    question = state["question"]

    answer = "Short summary for: " + question

    return {"answer": answer}
```

```python
def router(state: StudentState):
  question = state["question"].lower()
  if "quiz" in question:
    return "quiz_node"
  elif "summary" in question:
    return "summary_node"
  else:
    return "answer_node"
```

```python
graph_builder = StateGraph(StudentState)
```

```python
graph_builder.add_node("answer_node", answer_node)
graph_builder.add_node("quiz_node", quiz_node)
graph_builder.add_node("summary_node", summary_node)
```

```python
graph_builder.add_conditional_edges(
    START,
    router,
    {
        "answer_node": "answer_node",
        "quiz_node": "quiz_node",
        "summary_node": "summary_node",
    }
)
```

```python
graph_builder.add_edge("answer_node", END)
graph_builder.add_edge("quiz_node", END)
graph_builder.add_edge("summary_node", END)
```

```python
app = graph_builder.compile()
```

```python
result = app.invoke({
    "question": "Make quiz about LCEL"
})
```
